```python
import pandas as pd
import numpy as np

dados = {
    'Nome': ['Alice', 'Bob', 'Carlos', 'Diana', 'Eva', 'Fernando', 'Gustavo', 'Helena', 'Igor', 'Julia', 'Karine', 'Luiz', 'Mariana', 'Natalia', 'Oscar'],
    'Idade': [25, 30, 22, np.nan, 27, 34, 29, 28, np.nan, 31, 26, 33, 24, 29, 35],
    'Salario': [4000, 3500, 2800, 4200, 3200, np.nan, 3700, 2900, 4500, 3900, 4100, 3700, 3000, 3800, 3200],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Brasília', 'Curitiba', 'Porto Alegre', 'Salvador', 'Recife', 'Fortaleza', 'Manaus', 'Belém', 'Florianópolis', 'Vitória', 'Natal', 'Campo Grande']
}

df = pd.DataFrame(dados)


df.to_csv('dados.csv', index=False)


df = pd.read_csv('dados.csv')


print("\nVisualização das primeiras 5 linhas:")
print(df.head())


print("\nVisualização das últimas 5 linhas:")
print(df.tail())

print("\nAcessando um elemento específico usando iloc:")
print(df.iloc[2, 1])  # Linha 2, Coluna 1 (idade de Carlos)

print("\nAcessando um elemento específico usando loc:")
print(df.loc[3, 'Nome'])  # Linha 3, Coluna 'Nome' (Nome da Diana)

print("\nOrdenando os valores por idade em ordem crescente:")
df_idade_ordenada = df.sort_values(by='Idade')
print(df_idade_ordenada)

print("\nVerificando os valores nulos (dados faltantes):")
print(df.isnull())

print("\nDescrição estatística do DataFrame:")
print(df.describe())

print("\nNúmero de linhas e colunas do DataFrame:")
print("Número de linhas:", df.shape[0])
print("Número de colunas:", df.shape[1])

print("\nTipos de dados das colunas:")
print(df.dtypes)
```
