```python
numeros = []

for i in range(5):
    numero = int(input(f"Digite o {i+1}º número inteiro: "))
    numeros.append(numero)

soma = 0

for numero in numeros:
    soma += numero

print(f"A soma dos números digitados é: {soma}")
```

```python
alunos = {}

for i in range(3):
    nome = input(f"Digite o nome do aluno {i+1}: ")
    nota = float(input(f"Digite a nota do aluno {i+1}: "))
    alunos[nome] = nota

soma_notas = 0

for nome, nota in alunos.items():
    soma_notas += nota

media = soma_notas / len(alunos)

print(f"A média das notas é: {media}")
```

```python
local1 = {"Praia", "Montanha", "Floresta", "Cidade"}
local2 = {"Praia", "Deserto", "Cidade", "Campo"}

uniao = local1.union(local2)
print("União dos locais:", uniao)

intersecao = local1.intersection(local2)
print("Interseção dos locais:", intersecao)

diferenca = local1.difference(local2)
print("Diferença entre local1 e local2:", diferenca)

diferenca_simetrica = local1.symmetric_difference(local2)
print("Diferença simétrica entre local1 e local2:", diferenca_simetrica)
```
