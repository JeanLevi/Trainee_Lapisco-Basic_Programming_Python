```python
def remover_vogais(string):
    vogais = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
    string_sem_vogais = ''

    for caractere in string:
        if caractere not in vogais:
            string_sem_vogais += caractere

    return string_sem_vogais

entrada = input("Digite uma string: ")
resultado = remover_vogais(entrada)
print("String sem vogais:", resultado)
```

```python
def filtrar_palavras(string, lista_palavras):
    palavras_filtradas = []

    for palavra in lista_palavras:
        if palavra.lower() in string.lower():
            palavras_filtradas.append(palavra)

    for palavra in palavras_filtradas:
        string = string.replace(palavra, '')

    return string.strip()

entrada_string = input("Digite uma string: ")
entrada_lista = input("Digite uma lista de palavras separadas por espaço: ").split()

resultado = filtrar_palavras(entrada_string, entrada_lista)
print("String com as palavras filtradas:", resultado)
```
